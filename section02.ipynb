# Constants

```python
tensor = tf.constant([[23, 4], [32, 51]])
tensor
```

```python
tensor.shape
```

```python
tensor.numpy()
```

```python
np_array = np.array([[23, 4], [32, 51]])
```

```python
tensor_fromnp = tf.constant(np_array)
tensor_fromnp
```

# Variables

```python
tf_var = tf.Variable([[1., 2., 3.], [4., 5., 6.]])
tf_var
```

```python
tf_var[0, 2].assign(100)
tf_var
```

# Operations

```python
tensor = tf.constant([[1, 2], [3, 4]])
tensor
```

```python
tensor + 2
```

```python
tensor * 5
```

```python
np.square(tensor)
```

```python
np.sqrt(tensor)
```

```python
np.dot(tensor, tensor)
```

# Strings

```python
tf_str = tf.constant("Tensorflow")
tf_str
```

```python
tf.strings.length(tf_str)
```

```python
tf.strings.unicode_decode(tf_str, "UTF8")
```

```python
tf_str_array = tf.constant(["Tensorflow", "Deep Learning", "AI"])
tf_str_array
```

```python
for string in tf_str_array:
  print(string)
```
